```python
# helpful to visualize ordered pair of number of the form (x,y) as points inplane
```

```python
import matplotlib.pyplot as plt
```

```python
x_values = [1,2,3,4,5]
y_values = [1,4,9,16,25]
plt.scatter(x_values, y_values)
```

```python
# marker = set the shape of the dot
# c = set the color of the dots
```

```python
plt.scatter(x_values, y_values, marker='x', c='green')
```

```python
# Line Chart
```

```python
# plt.plot(), plots a line chart
plt.plot(x_values, y_values)
```

```python
# Drawing to Scale
```

```python
# if you want to see the graph to the scale, set x and y bounds to be the same
# still one unit on the x axis and one unit on the y axis won't be the same, till we make the graph a square by 
# setting set_size_inches(,)
plt.ylim(0,5)
plt.xlim(0,5)
plt.gcf().set_size_inches(5,5) # get current figure
```

```python
# you can set the title by setting plt.title and plt.xlabel and plt.ylabel
```

```python
def plot_segment(p1, p2):
    x1,y1 = p1
    x2, y2 = p2
    plt.plot([x1,x2], [y1,y2], marker='o')
```

```python
point1 = (0,3)
point2 = (2,1)
plot_segment(point1, point2)
```

```python
# drawing the sin function
```

```python
import numpy as np
plt.title('sin function')
plt.xlabel('number')
plt.ylabel('sin value')
xvalues = np.arange(0,10,0.1)
yvalues = np.sin(xvalues)

#set the scale
plt.xlimit = 10
plt.ylimit = 10
plt.gcf().set_size_inches(5,5)

plt.plot(xvalues,yvalues, c ='red', marker='o')
```
